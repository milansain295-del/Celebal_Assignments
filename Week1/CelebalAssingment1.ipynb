```python
import pandas as pd
import numpy as np

# Unzip the file into the current directory
!unzip -o archive.zip

# Load the dataset from the current directory (not from an 'archive/' folder)
df = pd.read_csv("Combined_dataset.csv")
print("Dataset loaded successfully!")
print("Initial Shape:", df.shape)
```

```python
# View the first 5 rows
print("--- FIRST 5 ROWS ---")
print(df.head())

# View the last 5 rows
print("\n--- LAST 5 ROWS ---")
print(df.tail())

print("\n--- DATASET SHAPE ---")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")

print("\n--- COLUMNS ---")
print(df.columns.tolist())

print("\n--- DATA TYPES & INFO ---")
print(df.info())
```

```python
print("--- PRE-CLEANING NULL COUNTS ---")
print(df.isnull().sum()[df.isnull().sum() > 0])

df['discount'] = df['discount'].fillna(0.0)

text_fill_cols = ['what_customers_said', 'seller_name', 'videos', 'variations', 'seller_information']
for col in text_fill_cols:
    df[col] = df[col].fillna("Not Available")

print("\nPost-cleaning overall null count:", df.isnull().sum().sum())
```

```python
premium_items = df[df['initial_price'] > 5000]
print(f"Filtered Data: Found {len(premium_items)} premium products above ₹5,000.")

subset_df = df[['product_id', 'title', 'category', 'initial_price', 'rating']]
print("\n--- SUBSET VIEW ---")
print(subset_df.head())
```

```python
duplicate_count = df.duplicated(subset=['product_id']).sum()
print(f"Detected duplicate entries: {duplicate_count}")

df = df.drop_duplicates(subset=['product_id'], keep='first')
print(f"Shape after duplicate management: {df.shape}")
```

```python
df['price'] = df['initial_price'].astype(float)

np.random.seed(42)
df['quantity'] = np.random.randint(1, 5, size=len(df))

df['total_amount'] = df['price'] * df['quantity']

print("--- VERIFYING DERIVED COLUMN OUTPUT ---")
print(df[['title', 'price', 'quantity', 'total_amount']].head())
```

```python
essential_columns = [
    'product_id',
    'title',
    'category',
    'rating',
    'price',
    'quantity',
    'total_amount'
]


df_lean = df[essential_columns]

print("--- LEAN DATASET PREVIEW ---")
print(df_lean.head())
print(f"\nNew Dataset Shape: {df_lean.shape[0]} rows, {df_lean.shape[1]} columns")

output_lean_filename = "lean_shopping_dataset.csv"
df_lean.to_csv(output_lean_filename, index=False)
print(f"\nSuccess! Streamlined dataset saved as: {output_lean_filename}")
```
